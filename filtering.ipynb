```python
people = {
    "first": ["Jane", "John"],
    "last": ["Doe", "Dave"],
    "email": ["janedoe@gmail.com", "johndave@gmail.com"]
}
```

```python
df1 = pd.DataFrame(people)
df1
```

```python
df1['last'] == 'Doe'

## True values met the filtering criteria and False values are those that didn't
## basically a mask is created over the entire dataframe with the logic
## then this mask can be applied on the dataframe to select required data
```

```python
df1_filt = (df1['last'] == 'Doe')
## so filter var is equal to a series of true false values
```

```python
df1[df1_filt]
```

```python
## same result as above
## loc is usually used to select rows by labels but if we pass in a series of booleans can also used to filter data
df1.loc[df1_filt]
```

```python
## with .loc can still grab specific columns
df1.loc[df1_filt, 'email']
```

```python
filt1 = (df1['last'] == 'Doe') & (df1['first'] == 'Jane')
```

```python
df1.loc[filt1, 'email']
```

```python
filt2 = (df1['last'] == 'Dave') | (df1['first'] == 'John')
```

```python
df1.loc[filt2, 'email']
```

```python
## to get the opposite of filter, use tilda
df1.loc[~filt2, 'email']
```
